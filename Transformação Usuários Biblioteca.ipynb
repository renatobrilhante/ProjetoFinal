# Base de usuários da Biblioteca online

Carregar o relatório de usuários da biblioteca, restringir os usuários que com mais de 60 dias sem acesso, trazer o status da matrícula dos alunos e identificar quais dos inativos ainda estão matriculados e seus cursos, para tratativas futuras. 


```python
import pandas as pd
import numpy as np
from datetime import datetime

#Caso ainda não tenha instalado, !pip install openpyxl
```

```python
df_b = pd.read_excel('Usuarios Biblioteca Virtual.xlsx')
df_b.head()
```

```python
df_a = pd.read_excel('Alunos.xlsx')
df_a.head()
```

#Tratando a df_a

```python
df_a.info()
```

```python
df_a['NOME_ALUNO'] = df_a['NOME_ALUNO'].str.title()
df_a['EMAIL'] = df_a['EMAIL'].str.lower()
df_a['CURSO'] = df_a['CURSO'].str.title()
df_a['NIVEL_ENSINO'] = df_a['NIVEL_ENSINO'].str.title()
df_a['STATUS_PERI_LETIVO'] = df_a['STATUS_PERI_LETIVO'].str.title()
```

```python
df_a.head()
```

```python
df_a['STATUS_PERI_LETIVO'].unique()
```

```python
Coluna = {'NOME_ALUNO':'Nome', 'EMAIL':'Email', 'CURSO':'Curso', 'RA':'R.A.', 'NIVEL_ENSINO':'Nível de Ensino', 'STATUS_PERI_LETIVO':'Status Matrícula'}
df_a.rename(columns=Coluna, inplace=True)
```

```python
df_a['Nível de Ensino'].unique()
```

```python
df_a.drop(['Nível de Ensino','R.A.'],axis=1,inplace=True)
```

```python
df_a
```

#Tratando a df_b

```python
df_b.head()
```

```python
df_b.info()
```

```python
df_b.rename(columns={'Id Pessoa':'ID Pessoa'}, inplace=True)
```

```python
df_b['Data Cadastro'] = pd.to_datetime(df_b['Data Cadastro'],format = '%d/%m/%Y %H:%M:%S')
```

```python
#ERROR!! df_b['Último Acesso'].astype('datetime64')
```

```python
df_b['Último Acesso2'] = df_b['Último Acesso'].str.replace('Sem Acesso',df_b['Último Acesso'].min(), regex=False)
```

```python
df_b['Último Acesso'] = df_b['Último Acesso'].str.replace('Sem Acesso','', regex=False)
df_b['Último Acesso'] = pd.to_datetime(df_b['Último Acesso'])
```

```python
df_b['Último Acesso2'] = pd.to_datetime(df_b['Último Acesso2'],format = '%d/%m/%Y %H:%M:%S')
df_b.head()
```

```python
df_b.info()
```

```python
df_b['Status2'] = df_b['Último Acesso2'].map(lambda x: 'Inativo' if abs((datetime.now()-x).days) > 60 else 'Ativo')
```

```python
df_b.head()
```

```python
df_b['Dias desde o último acesso'] = datetime.now()-df_b['Último Acesso2']
```

```python
df_b.head(30)
```

```python
df_b.drop(['Dias desde o último acesso','Último Acesso2','Status','Nome'],axis=1,inplace=True)
```

```python
df_b.rename(columns={'Status2':'Status'},inplace=True)
df_b.head()
```

```python
df= pd.merge(df_b,df_a,how='inner', on='Email')
```

```python
df.info()
```

```python
df.to_excel('Usuarios Biblioteca Virtual_atual.xlsx', index=False)
```

```python
filtro_inativos = df['Status'] == 'Inativo'

print('{} pessoas estão inativas na base.' .format(df[filtro_inativos]['Nome'].count()))
```

```python
cursos = df[filtro_inativos].groupby(['Curso'])
```

```python
#cursos que mais tem alunos inativos
```

```python
cursos.size().sort_values(ascending=False).head(10)
```

```python
status = df[filtro_inativos].groupby(['Status Matrícula'])
```

```python
#Número de alunos inativos por Status da Matrícula
```

```python
status.size().sort_values(ascending=False)
```
